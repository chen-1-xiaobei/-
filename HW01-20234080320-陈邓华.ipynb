### 3.1 理论计算题
某疾病在人群中的患病率为 0.1%。现有一种检测方法：
- 若患病，检测呈阳性的概率为 99%（灵敏度）；
- 若未患病，检测呈阳性的概率为 2%（假阳性率）。
一个人检测结果为阳性，求他真正患病的概率（使用贝叶斯公式）。

```python
# 使用贝叶斯公式计算
# P(A|B) = P(B|A) * P(A) / P(B)

# 定义已知概率
P_A = 0.001  # 患病率 P(A)
P_not_A = 1 - P_A  # 未患病率 P(¬A)
P_B_given_A = 0.99  # 患病时检测呈阳性的概率 P(B|A)
P_B_given_not_A = 0.02  # 未患病时检测呈阳性的概率 P(B|¬A)

# 计算全概率 P(B)
P_B = P_B_given_A * P_A + P_B_given_not_A * P_not_A

# 计算后验概率 P(A|B)
P_A_given_B = P_B_given_A * P_A / P_B

print(f'3.1 理论计算题：')
print(f'一个人检测结果为阳性，他真正患病的概率 = {P_A_given_B:.4f}')
```

### 3.2 编程题
用 Python 模拟中心极限定理：
1. 生成 n = 10000 个服从均匀分布 U(0, 1) 的随机变量，重复 m = 1000 次，每次计算这 n 个变量的均值。
2. 绘制这 1000 个均值的直方图，并叠加理论正态分布的概率密度曲线。
3. 计算这些均值的实际方差。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 设置参数
n = 10000  # 每次生成的随机变量个数
m = 1000   # 重复次数

# 1. 生成随机变量并计算均值
means = []
for _ in range(m):
    # 生成 n 个服从均匀分布 U(0, 1) 的随机变量
    samples = np.random.uniform(0, 1, n)
    # 计算均值并添加到列表
    means.append(np.mean(samples))

# 转换为 numpy 数组
means = np.array(means)

# 3. 计算这些均值的实际方差
actual_variance = np.var(means)
print(f'3. 这些均值的实际方差 = {actual_variance:.6f}')

# 2. 绘制直方图并叠加理论正态分布的概率密度曲线
plt.figure(figsize=(10, 6))

# 绘制直方图
plt.hist(means, bins=30, density=True, alpha=0.6, color='g', label='样本均值分布')

# 计算理论正态分布的参数
# 均匀分布 U(0, 1) 的均值为 0.5，方差为 1/12
# 根据中心极限定理，样本均值的分布近似为 N(0.5, (1/12)/n)
mu = 0.5
sigma = np.sqrt((1/12) / n)

# 生成理论正态分布的 x 值
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# 计算理论正态分布的概率密度
pdf = norm.pdf(x, mu, sigma)

# 绘制理论正态分布的概率密度曲线
plt.plot(x, pdf, 'r-', lw=2, label='理论正态分布')

# 添加标题和标签
plt.title('中心极限定理模拟：均匀分布样本均值的分布')
plt.xlabel('均值')
plt.ylabel('概率密度')
plt.legend()
plt.grid(True)

# 显示图形
plt.show()
```

## 4 导数、反向传播和复杂度

### 4.1 理论计算题
给定函数 z = (w₁x₁ + w₂x₂ − y)²，其中 x₁ = 2, x₂ = 1, y = 3 为常数，w₁, w₂ 为变量。
1. 分别计算 ∂z/∂w₁ 和 ∂z/∂w₂ 的表达式。
2. 若当前 w₁ = 0.5, w₂ = 1，计算梯度值。

```python
# 4.1 理论计算题
print('4.1 理论计算题：')
print('1. 偏导数表达式：')
print('   ∂z/∂w₁ = 2(w₁x₁ + w₂x₂ - y) * x₁ = 2(2w₁ + w₂ - 3) * 2 = 4(2w₁ + w₂ - 3)')
print('   ∂z/∂w₂ = 2(w₁x₁ + w₂x₂ - y) * x₂ = 2(2w₁ + w₂ - 3) * 1 = 2(2w₁ + w₂ - 3)')

# 2. 计算梯度值
w1 = 0.5
w2 = 1
x1 = 2
x2 = 1
y = 3

# 计算中间值
inner = w1 * x1 + w2 * x2 - y

# 计算偏导数
dz_dw1 = 2 * inner * x1
dz_dw2 = 2 * inner * x2

print('2. 梯度值：')
print(f'   当 w₁ = {w1}, w₂ = {w2} 时：')
print(f'   ∂z/∂w₁ = {dz_dw1}')
print(f'   ∂z/∂w₂ = {dz_dw2}')
```

### 4.2 编程题
手动实现一个简单计算图的反向传播：
1. 定义前向计算：输入 x，经过 a = x · w₁，b = a + w₂，输出 L = b²。
2. 给定 x = 2, w₁ = 1.5, w₂ = 0.5，计算前向结果。
3. 手动（按链式法则）计算损失 L 对 w₁, w₂ 的梯度，并验证与自动微分工具的结果是否一致。

```python
import torch

# 4.2 编程题
# 1. 定义前向计算
# 2. 给定参数值
x = 2
w1 = 1.5
w2 = 0.5

# 前向计算
a = x * w1
b = a + w2
L = b ** 2

print('4.2 编程题：')
print(f'前向计算结果：')
print(f'   a = x · w₁ = {a}')
print(f'   b = a + w₂ = {b}')
print(f'   L = b² = {L}')

# 3. 手动计算梯度（链式法则）
# ∂L/∂b = 2b
# ∂b/∂a = 1, ∂b/∂w₂ = 1
# ∂a/∂w₁ = x
# ∂L/∂w₁ = ∂L/∂b * ∂b/∂a * ∂a/∂w₁ = 2b * 1 * x
# ∂L/∂w₂ = ∂L/∂b * ∂b/∂w₂ = 2b * 1

dL_db = 2 * b
dL_dw1 = dL_db * 1 * x
dL_dw2 = dL_db * 1

print(f'手动计算梯度：')
print(f'   ∂L/∂w₁ = {dL_dw1}')
print(f'   ∂L/∂w₂ = {dL_dw2}')

# 使用 PyTorch 自动微分验证
x_torch = torch.tensor(x, requires_grad=False)
w1_torch = torch.tensor(w1, requires_grad=True)
w2_torch = torch.tensor(w2, requires_grad=True)

# 前向计算
a_torch = x_torch * w1_torch
b_torch = a_torch + w2_torch
L_torch = b_torch ** 2

# 反向传播
L_torch.backward()

print(f'PyTorch 自动微分结果：')
print(f'   ∂L/∂w₁ = {w1_torch.grad.item()}')
print(f'   ∂L/∂w₂ = {w2_torch.grad.item()}')

# 验证结果是否一致
print(f'验证结果：')
print(f'   ∂L/∂w₁ 结果一致：{abs(dL_dw1 - w1_torch.grad.item()) < 1e-10}')
print(f'   ∂L/∂w₂ 结果一致：{abs(dL_dw2 - w2_torch.grad.item()) < 1e-10}')
```

## 5 线性方法、基础优化和 softmax 回归

### 5.1 理论计算题
考虑线性回归模型 y = wx + b，使用均方误差损失 L = (1/n) ∑(yᵢ − (wxᵢ + b))²
推导损失对 w 和 b 的偏导数表达式（用求和形式）。

∂L/∂w = –(2/n) Σ_{i=1}^n x_i (y_i – w x_i – b)
∂L/∂b = –(2/n) Σ_{i=1}^n (y_i – w x_i – b)

### 5.2 编程题
从零实现 softmax 回归对手写数字（MNIST）进行分类：
1. 使用 sklearn.datasets.load_digits() 或自己加载 MNIST 子集。
2. 将图像展平为向量，对标签进行 one‑hot 编码。
3. 实现 softmax 函数、交叉熵损失函数。
4. 使用小批量随机梯度下降（批量大小 32，学习率 0.1）训练模型 50 个 epoch。
5. 输出测试集上的准确率。

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

# 5.2 编程题：从零实现 softmax 回归

# 1. 加载数据集
digits = load_digits()
X = digits.data  # 图像数据
y = digits.target  # 标签

# 数据预处理
# 归一化
X = X / 16.0  # 像素值范围为 0-16，归一化到 0-1

# 2. 将图像展平为向量（已经是展平的，每个图像是 8x8=64 维向量）
# 对标签进行 one-hot 编码
def one_hot_encode(y, num_classes):
    n = y.shape[0]
    one_hot = np.zeros((n, num_classes))
    one_hot[np.arange(n), y] = 1
    return one_hot

num_classes = 10  # 数字 0-9
y_one_hot = one_hot_encode(y, num_classes)

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, test_size=0.2, random_state=42)

# 3. 实现 softmax 函数
def softmax(z):
    # 防止数值溢出
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

# 实现交叉熵损失函数
def cross_entropy_loss(y_pred, y_true):
    n = y_true.shape[0]
    # 避免 log(0)
    epsilon = 1e-12
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.sum(y_true * np.log(y_pred)) / n

# 4. 初始化参数
input_dim = X_train.shape[1]  # 输入维度，64
output_dim = num_classes  # 输出维度，10

# 随机初始化权重和偏置
np.random.seed(42)
W = np.random.randn(input_dim, output_dim) * 0.01
b = np.zeros(output_dim)

# 训练参数
batch_size = 32
learning_rate = 0.1
epochs = 50

# 训练模型
for epoch in range(epochs):
    # 打乱训练数据
    permutation = np.random.permutation(X_train.shape[0])
    X_train_shuffled = X_train[permutation]
    y_train_shuffled = y_train[permutation]
    
    # 小批量训练
    for i in range(0, X_train.shape[0], batch_size):
        # 取一个批次的数据
        batch_X = X_train_shuffled[i:i+batch_size]
        batch_y = y_train_shuffled[i:i+batch_size]
        
        # 前向传播
        z = np.dot(batch_X, W) + b
        y_pred = softmax(z)
        
        # 计算损失
        loss = cross_entropy_loss(y_pred, batch_y)
        
        # 反向传播计算梯度
        grad_z = y_pred - batch_y
        grad_W = np.dot(batch_X.T, grad_z) / batch_size
        grad_b = np.sum(grad_z, axis=0) / batch_size
        
        # 更新参数
        W -= learning_rate * grad_W
        b -= learning_rate * grad_b
    
    # 计算训练集准确率
    z_train = np.dot(X_train, W) + b
    y_pred_train = softmax(z_train)
    y_pred_train_labels = np.argmax(y_pred_train, axis=1)
    y_train_labels = np.argmax(y_train, axis=1)
    train_accuracy = np.mean(y_pred_train_labels == y_train_labels)
    
    # 计算测试集准确率
    z_test = np.dot(X_test, W) + b
    y_pred_test = softmax(z_test)
    y_pred_test_labels = np.argmax(y_pred_test, axis=1)
    y_test_labels = np.argmax(y_test, axis=1)
    test_accuracy = np.mean(y_pred_test_labels == y_test_labels)
    
    if (epoch + 1) % 10 == 0:
        print(f'Epoch {epoch+1}/{epochs}, Loss: {loss:.4f}, Train Accuracy: {train_accuracy:.4f}, Test Accuracy: {test_accuracy:.4f}')

# 5. 输出测试集上的准确率
print(f'最终测试集准确率：{test_accuracy:.4f}')
```

## 6 最大似然估计和逻辑回归

### 6.1 理论计算题
假设样本 x₁, x₂, ..., xₙ 独立同分布于正态分布 N(µ, σ²)。
1. 写出似然函数 L(µ, σ²)。
2. 证明：µ 的最大似然估计为 µ̂ = (1/n) ∑xᵢ。
3. 证明：σ² 的最大似然估计为 σ̂² = (1/n) ∑(xᵢ − µ̂)²。

1.似然函数
L(μ, σ²) = ∏{i=1}^{n} (1 / √(2πσ²)) exp[-(x_i - μ)² / (2σ²)]
= (2πσ²)^{-n/2} exp[ -∑{i=1}^{n} (x_i - μ)² / (2σ²) ]

2.取对数似然：ℓ = ln L = -n/2 ln(2π) - n/2 ln(σ²) - (1/(2σ²)) ∑(x_i - μ)²
对 μ 求偏导：∂ℓ/∂μ = - (1/(2σ²)) * (-2)∑(x_i - μ) = (1/σ²) ∑(x_i - μ)
令 ∂ℓ/∂μ = 0 ⇒ ∑(x_i - μ) = 0 ⇒ ∑x_i - nμ = 0 ⇒ μ = (1/n)∑x_i
二阶导数为负，故为极大值点。因此 μ̂ = (1/n)∑x_i。

3.将 σ² 视为参数（记 θ = σ²），对数似然对 θ 求导：
∂ℓ/∂θ = -n/(2θ) + (1/(2θ²)) ∑(x_i - μ)²
令其为零：-n/(2θ) + (1/(2θ²))∑(x_i - μ)² = 0
两边乘以 2θ²：-nθ + ∑(x_i - μ)² = 0 ⇒ θ = (1/n)∑(x_i - μ)²
此时 μ 用其 MLE 估计 μ̂ 代入，即得 σ̂² = (1/n)∑(x_i - μ̂)²。
二阶导数验证可知为极大值。

### 6.2 编程题
实现二分类逻辑回归模型（不使用深度学习框架的高阶 API）：
1. 生成两类线性可分的数据：每类 200 个样本，特征维度为 2。
2. 定义 sigmoid 函数和二元交叉熵损失。
3. 使用梯度下降法（学习率 0.1，迭代 1000 次）学习参数 w, b。
4. 绘制决策边界，并计算模型在测试集（新生成 100 个样本）上的分类准确率。

```python
import numpy as np
import matplotlib.pyplot as plt

# 6.2 编程题：实现二分类逻辑回归模型

# 1. 生成两类线性可分的数据
np.random.seed(42)

# 类别 0：均值为 [0, 0]，协方差矩阵为 [[1, 0], [0, 1]]
class0 = np.random.randn(200, 2) + np.array([0, 0])
y0 = np.zeros(200)

# 类别 1：均值为 [3, 3]，协方差矩阵为 [[1, 0], [0, 1]]
class1 = np.random.randn(200, 2) + np.array([3, 3])
y1 = np.ones(200)

# 合并数据
X = np.vstack([class0, class1])
y = np.hstack([y0, y1])

# 2. 定义 sigmoid 函数
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# 定义二元交叉熵损失
def binary_cross_entropy_loss(y_pred, y_true):
    epsilon = 1e-12
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# 3. 使用梯度下降法学习参数 w, b
# 初始化参数
input_dim = 2
w = np.random.randn(input_dim) * 0.01
b = 0

# 训练参数
learning_rate = 0.1
iterations = 1000

# 训练模型
for i in range(iterations):
    # 前向传播
    z = np.dot(X, w) + b
    y_pred = sigmoid(z)
    
    # 计算损失
    loss = binary_cross_entropy_loss(y_pred, y)
    
    # 计算梯度
    grad_w = np.dot(X.T, (y_pred - y)) / len(y)
    grad_b = np.mean(y_pred - y)
    
    # 更新参数
    w -= learning_rate * grad_w
    b -= learning_rate * grad_b
    
    if (i + 1) % 100 == 0:
        print(f'Iteration {i+1}/{iterations}, Loss: {loss:.4f}')

# 4. 绘制决策边界
plt.figure(figsize=(10, 6))

# 绘制数据点
plt.scatter(class0[:, 0], class0[:, 1], c='blue', label='Class 0')
plt.scatter(class1[:, 0], class1[:, 1], c='red', label='Class 1')

# 绘制决策边界（sigmoid(z) = 0.5 时，z = 0）
x1 = np.linspace(-3, 6, 100)
x2 = (-w[0] * x1 - b) / w[1]
plt.plot(x1, x2, 'k-', label='Decision Boundary')

# 添加标题和标签
plt.title('Logistic Regression Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)

# 显示图形
plt.show()

# 生成测试集数据
test_class0 = np.random.randn(50, 2) + np.array([0, 0])
test_y0 = np.zeros(50)
test_class1 = np.random.randn(50, 2) + np.array([3, 3])
test_y1 = np.ones(50)

X_test = np.vstack([test_class0, test_class1])
y_test = np.hstack([test_y0, test_y1])

# 计算测试集准确率
z_test = np.dot(X_test, w) + b
y_pred_test = sigmoid(z_test)
y_pred_test_labels = (y_pred_test >= 0.5).astype(int)
test_accuracy = np.mean(y_pred_test_labels == y_test)

print(f'测试集准确率：{test_accuracy:.4f}')
```
